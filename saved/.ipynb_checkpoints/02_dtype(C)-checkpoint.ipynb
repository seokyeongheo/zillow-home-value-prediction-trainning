```python
def find_others(pvalue, a, b):
    others = []
    length = len(pvalue)

    for i in range(length):
        if pvalue.values[i] > 0.01:
            others.append(pvalue.index[i][a:b])

    del others[0]
    del others[-1]
    return others
```
