```python
# poolsizesum, poolcnt
poolsizesum_median = df.loc[df['poolcnt'] > 0, 'poolsizesum'].median()
df.loc[(df['poolcnt'] > 0) & (df['poolsizesum'].isnull()), 'poolsizesum'] = poolsizesum_median
```

```python
# 수영장 없는 곳은 사이즈 0으로 채우기
df.loc[(df['poolcnt'].isnull()), 'poolsizesum'] = 0
```

```python
# 수영장 NaN 인 경우 0으로 채우기
df.loc[(df['poolcnt'].isnull()), 'poolcnt'] = 0
```

```python
# 핫텁이나 스파 없는 곳은 0으로 채우기, hashottuborspa 있는 곳 True 값 1로 채우기
df['hashottuborspa']= np.where(df['hashottuborspa']== "TRUE",1,0)
```
